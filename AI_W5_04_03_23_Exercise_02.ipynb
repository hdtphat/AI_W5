```python
# cài fuzzy logic
!pip install -U scikit-fuzzy
# thêm thư viện
import numpy as np
import skfuzzy as fz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

```python
# biến vào (windspeed, temperature)
fever = ctrl.Antecedent(np.arange(0,1.1,0.1),'fever')
headache = ctrl.Antecedent(np.arange(0,1.1,0.1),'headache')
nausea = ctrl.Antecedent(np.arange(0,1.1,0.1),'nausea')
vomiting = ctrl.Antecedent(np.arange(0,1.1,0.1),'vomiting')
jaundice = ctrl.Antecedent(np.arange(0,1.1,0.1),'jaundice')
enlarge_liver = ctrl.Antecedent(np.arange(0,1.1,0.1),'enlarge_liver')
joint_pain = ctrl.Antecedent(np.arange(0,1.1,0.1),'joint_pain')
body_weakness = ctrl.Antecedent(np.arange(0,1.1,0.1),'body_weakness')
dizziness = ctrl.Antecedent(np.arange(0,1.1,0.1),'dizziness')
loss_of_appetite = ctrl.Antecedent(np.arange(0,1.1,0.1),'loss_of_appetite')
mp = ctrl.Antecedent(np.arange(0,1.1,0.1),'mp')

# biến ra (rainfall)
conclusion = ctrl.Consequent(np.arange(0,1.1,0.1),'conclusion')

# tạo các giá trị cho bảng
# mi, mo, se, vs (mild, moderate, severe, very severe)
fever['mi'] = fz.trimf(fever.universe, [0,0.2,0.4])
fever['mo'] = fz.trimf(fever.universe, [0.2,0.3,0.6])
fever['se'] = fz.trimf(fever.universe, [0.4,0.5,0.8])
fever['vs'] = fz.trimf(fever.universe, [0.6,0.9,1])
fever.view()

headache['mi'] = fz.trimf(headache.universe, [0,0.3,0.4])
headache['mo'] = fz.trimf(headache.universe, [0.2,0.4,0.6])
headache['se'] = fz.trimf(headache.universe, [0.4,0.7,0.8])
headache['vs'] = fz.trimf(headache.universe, [0.6,0.8,1])
headache.view()

nausea['mi'] = fz.trimf(nausea.universe, [0,0.3,0.4])
nausea['mo'] = fz.trimf(nausea.universe, [0.2,0.5,0.6])
nausea['se'] = fz.trimf(nausea.universe, [0.4,0.6,0.8])
nausea['vs'] = fz.trimf(nausea.universe, [0.6,0.8,1])
nausea.view()

vomiting['mi'] = fz.trimf(vomiting.universe, [0,0.3,0.4])
vomiting['mo'] = fz.trimf(vomiting.universe, [0.2,0.4,0.6])
vomiting['se'] = fz.trimf(vomiting.universe, [0.4,0.7,0.8])
vomiting['vs'] = fz.trimf(vomiting.universe, [0.6,0.9,1])
vomiting.view()

jaundice['mi'] = fz.trimf(jaundice.universe, [0,0.3,0.4])
jaundice['mo'] = fz.trimf(jaundice.universe, [0.2,0.4,0.6])
jaundice['se'] = fz.trimf(jaundice.universe, [0.4,0.7,0.8])
jaundice['vs'] = fz.trimf(jaundice.universe, [0.6,0.9,1])
jaundice.view()

enlarge_liver['mi'] = fz.trimf(enlarge_liver.universe, [0,0.1,0.4])
enlarge_liver['mo'] = fz.trimf(enlarge_liver.universe, [0.2,0.3,0.6])
enlarge_liver['se'] = fz.trimf(enlarge_liver.universe, [0.4,0.5,0.8])
enlarge_liver['vs'] = fz.trimf(enlarge_liver.universe, [0.6,0.8,1])
enlarge_liver.view()

joint_pain['mi'] = fz.trimf(joint_pain.universe, [0,0.1,0.4])
joint_pain['mo'] = fz.trimf(joint_pain.universe, [0.2,0.3,0.6])
joint_pain['se'] = fz.trimf(joint_pain.universe, [0.4,0.7,0.8])
joint_pain['vs'] = fz.trimf(joint_pain.universe, [0.6,0.9,1])
joint_pain.view()

body_weakness['mi'] = fz.trimf(body_weakness.universe, [0,0.2,0.4])
body_weakness['mo'] = fz.trimf(body_weakness.universe, [0.2,0.3,0.6])
body_weakness['se'] = fz.trimf(body_weakness.universe, [0.4,0.4,0.8])
body_weakness['vs'] = fz.trimf(body_weakness.universe, [0.6,0.7,1])
body_weakness.view()

dizziness['mi'] = fz.trimf(dizziness.universe, [0,0.2,0.4])
dizziness['mo'] = fz.trimf(dizziness.universe, [0.2,0.4,0.6])
dizziness['se'] = fz.trimf(dizziness.universe, [0.4,0.6,0.8])
dizziness['vs'] = fz.trimf(dizziness.universe, [0.6,0.8,1])
dizziness.view()

loss_of_appetite['mi'] = fz.trimf(loss_of_appetite.universe, [0,0,0.4])
loss_of_appetite['mo'] = fz.trimf(loss_of_appetite.universe, [0.2,0.3,0.6])
loss_of_appetite['se'] = fz.trimf(loss_of_appetite.universe, [0.4,0.8,0.8])
loss_of_appetite['vs'] = fz.trimf(loss_of_appetite.universe, [0.6,0.7,1])
loss_of_appetite.view()

mp['mi'] = fz.trimf(mp.universe, [0,0.3,0.4])
mp['mo'] = fz.trimf(mp.universe, [0.2,0.6,0.6])
mp['se'] = fz.trimf(mp.universe, [0.4,0.7,0.8])
mp['vs'] = fz.trimf(mp.universe, [0.6,0.9,1])
mp.view()

conclusion['mi'] = fz.trimf(conclusion.universe, [0,0.2,0.4])
conclusion['mo'] = fz.trimf(conclusion.universe, [0.2,0.4,0.6])
conclusion['se'] = fz.trimf(conclusion.universe, [0.4,0.6,0.8])
conclusion['vs'] = fz.trimf(conclusion.universe, [0.6,0.8,1])
conclusion.view()

#tạo các rules
rule_1 = ctrl.Rule(fever['mi']& headache['mi']& nausea['mi']& vomiting['mi']& jaundice['mi']& enlarge_liver['mi']& joint_pain['mi']& body_weakness['mi']& dizziness['se']& loss_of_appetite['mi']& mp['mi'], conclusion['mi'])
rule_2 = ctrl.Rule(fever['mo']& headache['mi']& nausea['mi']& vomiting['mi']& jaundice['mi']& enlarge_liver['mi']& joint_pain['mo']& body_weakness['mo']& dizziness['se']& loss_of_appetite['se']& mp['mo'], conclusion['mo'])
rule_3 = ctrl.Rule(fever['se']& headache['mo']& nausea['mi']& vomiting['mi']& jaundice['mi']& enlarge_liver['mi']& joint_pain['mi']& body_weakness['se']& dizziness['se']& loss_of_appetite['se']& mp['mo'], conclusion['se'])
rule_4 = ctrl.Rule(fever['vs']& headache['mi']& nausea['mi']& vomiting['mi']& jaundice['mi']& enlarge_liver['mi']& joint_pain['se']& body_weakness['se']& dizziness['mi']& loss_of_appetite['mi']& mp['mi'], conclusion['vs'])
rule_5 = ctrl.Rule(fever['mo']& headache['mi']& nausea['mi']& vomiting['mo']& jaundice['mi']& enlarge_liver['mi']& joint_pain['mo']& body_weakness['mo']& dizziness['mo']& loss_of_appetite['se']& mp['mo'], conclusion['mo'])

rule_6  = ctrl.Rule(fever['mi']& headache['mo']& nausea['mo']& vomiting['mi']& jaundice['mi']& enlarge_liver['mi']& joint_pain['mi']& body_weakness['mi']& dizziness['mo']& loss_of_appetite['mi']& mp['mi'], conclusion['mi'])
rule_7  = ctrl.Rule(fever['mi']& headache['mi']& nausea['mo']& vomiting['mo']& jaundice['mi']& enlarge_liver['mi']& joint_pain['se']& body_weakness['se']& dizziness['mo']& loss_of_appetite['mo']& mp['mo'], conclusion['se'])
rule_8  = ctrl.Rule(fever['mo']& headache['mo']& nausea['mo']& vomiting['mo']& jaundice['mo']& enlarge_liver['mo']& joint_pain['mo']& body_weakness['mo']& dizziness['mo']& loss_of_appetite['mo']& mp['mo'], conclusion['mo'])
rule_9  = ctrl.Rule(fever['mo']& headache['mi']& nausea['mo']& vomiting['mo']& jaundice['mi']& enlarge_liver['se']& joint_pain['mo']& body_weakness['mo']& dizziness['se']& loss_of_appetite['mo']& mp['mo'], conclusion['mo'])
rule_10 = ctrl.Rule(fever['mi']& headache['mi']& nausea['mo']& vomiting['mo']& jaundice['mi']& enlarge_liver['mo']& joint_pain['mi']& body_weakness['mi']& dizziness['mi']& loss_of_appetite['mi']& mp['mi'], conclusion['mi'])

rule_11 = ctrl.Rule(fever['se']& headache['se']& nausea['se']& vomiting['se']& jaundice['se']& enlarge_liver['se']& joint_pain['se']& body_weakness['vs']& dizziness['se']& loss_of_appetite['se']& mp['mo'], conclusion['vs'])
rule_12 = ctrl.Rule(fever['se']& headache['se']& nausea['mo']& vomiting['se']& jaundice['mo']& enlarge_liver['se']& joint_pain['mo']& body_weakness['se']& dizziness['mo']& loss_of_appetite['mi']& mp['mo'], conclusion['se'])
rule_13 = ctrl.Rule(fever['mo']& headache['mo']& nausea['mo']& vomiting['mo']& jaundice['mi']& enlarge_liver['mi']& joint_pain['mi']& body_weakness['mo']& dizziness['mi']& loss_of_appetite['mo']& mp['mo'], conclusion['mo'])
rule_14 = ctrl.Rule(fever['se']& headache['se']& nausea['mo']& vomiting['se']& jaundice['se']& enlarge_liver['se']& joint_pain['se']& body_weakness['se']& dizziness['mo']& loss_of_appetite['mo']& mp['se'], conclusion['se'])
rule_15 = ctrl.Rule(fever['mi']& headache['mi']& nausea['mi']& vomiting['mo']& jaundice['mi']& enlarge_liver['mi']& joint_pain['mi']& body_weakness['se']& dizziness['mi']& loss_of_appetite['se']& mp['mo'], conclusion['se'])

rule_16 = ctrl.Rule(fever['mo']& headache['mo']& nausea['mi']& vomiting['mo']& jaundice['se']& enlarge_liver['mo']& joint_pain['mi']& body_weakness['vs']& dizziness['se']& loss_of_appetite['mi']& mp['mo'], conclusion['vs'])
rule_17 = ctrl.Rule(fever['vs']& headache['vs']& nausea['mo']& vomiting['mo']& jaundice['mi']& enlarge_liver['mo']& joint_pain['mi']& body_weakness['mo']& dizziness['vs']& loss_of_appetite['mi']& mp['mo'], conclusion['vs'])
rule_18 = ctrl.Rule(fever['vs']& headache['vs']& nausea['vs']& vomiting['mi']& jaundice['se']& enlarge_liver['se']& joint_pain['mo']& body_weakness['se']& dizziness['vs']& loss_of_appetite['vs']& mp['se'], conclusion['vs'])
rule_19 = ctrl.Rule(fever['mo']& headache['mo']& nausea['mo']& vomiting['mo']& jaundice['mi']& enlarge_liver['mi']& joint_pain['mo']& body_weakness['mo']& dizziness['mo']& loss_of_appetite['mo']& mp['mo'], conclusion['mo'])
rule_20 = ctrl.Rule(fever['se']& headache['se']& nausea['se']& vomiting['se']& jaundice['se']& enlarge_liver['se']& joint_pain['se']& body_weakness['se']& dizziness['se']& loss_of_appetite['se']& mp['mo'], conclusion['vs'])

# dạy máy học
conclusion_calc = ctrl.ControlSystem([rule_1, rule_2, rule_3, rule_4, rule_5, rule_6, rule_7, rule_8, rule_9, rule_10, rule_11, rule_12, rule_13, rule_14, rule_15, rule_16, rule_17, rule_18, rule_19, rule_20])
conclusion_sim = ctrl.ControlSystemSimulation(conclusion_calc)

#cho input đầu vào
conclusion_sim.input['fever'] = 0.5
conclusion_sim.input['headache'] = 0.5
conclusion_sim.input['nausea'] = 0.5
conclusion_sim.input['vomiting'] = 0.5
conclusion_sim.input['jaundice'] = 0.5
conclusion_sim.input['enlarge_liver'] = 0.5
conclusion_sim.input['joint_pain'] = 0.5
conclusion_sim.input['body_weakness'] = 0.5
conclusion_sim.input['dizziness'] = 0.5
conclusion_sim.input['loss_of_appetite'] = 0.5
conclusion_sim.input['mp'] = 0.5

#tính toán
conclusion_sim.compute()

#hiển thị kết quả
print(conclusion_sim.output['conclusion'])
conclusion.view(sim = conclusion_sim)
```
